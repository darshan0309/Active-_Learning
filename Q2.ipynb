** Generate random dataframe

```python
random = np.random.randint(1,101,(100,4))
rdf = pd.DataFrame(random,columns = ['C1','C2','C3','C4'])
rdf
```

```python
c1mean = rdf.C1.mean()
c2mean = rdf.C2.mean()
c3mean = rdf.C3.mean()
c4mean = rdf.C4.mean()
print(c1mean)
print(c2mean)
print(c3mean)
print(c3mean)
```

**Filling the na values with the mean calculated above

```python
rdf.C1.fillna(c1mean)
rdf.C2.fillna(c2mean)
rdf.C3.fillna(c3mean)
rdf.C4.fillna(c4mean)
rdf
```

** Sorting wrt to first two columns

Question 9

** Generating two dataframe in which one column is common

```python
random1 = np.random.randint(1,10,(100,4))
dataframe1 = pd.DataFrame(random1,columns = ['commoncolumn','C2','C3','C4'])
random2 = np.random.randint(1,10,(100,4))
dataframe2 = pd.DataFrame(random2,columns = ['commoncolumn','D2','D3','D4'])
```

```python
dataframe1
```

```python
dataframe2
```

** Merging them based on the basis of there common column

```python
merged_df = pd.merge(dataframe1, dataframe2,on = 'commoncolumn')# here the value which the common column have in both the dataframe are printed if the value
merged_df                                                       # that occur multiple time is printed multiple times eg- if value 5 occurs 2 times
                                                                # in first one and 3 times in second one so there are total 6 possibilities(2*3 = 6)
                                                                # and will be printed 6 times in merged one if all are unique.
```
